```python
def play_game():
    board = chess.Board()
    while not(board.is_insufficient_material()) and not(board.is_checkmate()) and not(board.is_stalemate()):
        if board.ply() > 100:
            return('0-0')
        if board.turn == True:
            prediction = model_get_move(board, [1])
            board.push(prediction)
        else:
            board.push(rand_get_move(board))
    return(board.result())

outcome = []
for i in range(100):
    result = play_game()
    print(result)
    outcome.append(result)
```
